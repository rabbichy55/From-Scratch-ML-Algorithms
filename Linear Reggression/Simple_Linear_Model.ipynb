```python
# Built our own linear reggression model
class LinearReggresion_Model_:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X_train1, y_train1):

        numerator = 0
        denominator = 0
        
        for i in range(X_train1.shape[0]):
            
            numerator = numerator + ((X_train1[i] - X_train1.mean())*(y_train1[i] - y_train1.mean()))
            denominator = denominator + ((X_train1[i] - X_train1.mean())*(X_train1[i] - X_train1.mean()))
            
        self.m = numerator/denominator
        self.b = y_train1.mean() - (self.m * X_train1.mean())
        print(self.m) # slope
        print(self.b) # intercept

    def predict(self, X_test1):
        print(X_test1)
        return self.m * X_test1 + self.b
        
```
